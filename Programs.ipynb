```python
"""Task 1: Basic Line Plot
Create a DataFrame with columns:
day → 1 to 7
sales → random integers between 100 and 500
Plot a line chart showing sales trend across days."""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'days': [1, 2, 3, 4, 5, 6, 7],
    'sales': np.random.randint(100, 501, size = 7)
}
df = pd.DataFrame(data)
df

#Line chart
plt.plot(df['days'], df['sales'], color = 'red', marker = "*")
plt.title("Line Chart")
plt.xlabel("days")
plt.ylabel("sales")
plt.show()
```

```python
"""Task 2: Missing Values Bar Plot
Create a DataFrame with:
name
age (include some NaN values)
Fill missing ages with the mean.
 Plot a bar chart of name vs age."""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create DataFrame
df = pd.DataFrame({
    "name": ["Asha", "Ravi", "Sita", "Manoj", "Kiran"],
    "age": [22, np.nan, 24, np.nan, 26]
})

# Filling missing ages with mean
df["age"] = df["age"].fillna(df["age"].mean())

# Plot bar chart
plt.bar(df["name"], df["age"])
plt.xlabel("Name")
plt.ylabel("Age")
plt.title("Bar Chart of Name vs Age")
plt.show()
```

```python
"""Task 3: GroupBy + Bar Chart
Create a DataFrame with:
department
salary
Calculate average salary per department.
 Visualize using a bar chart."""

import pandas as pd
import matplotlib.pyplot as plt
data = {
    'department': ['HR', 'IT', 'Sales', 'HR'],
    'salary': [20000, 30000, 25000, 40000]
}
df = pd.DataFrame(data)
print(df)
print("\n")
avg_sal = (df.groupby('department')['salary'].mean().reset_index())
print("Average salary per department: \n",avg_sal)

#Bar Chart
print("\n")
plt.bar(df['department'], df['salary'], color = 'orange')
plt.show()
```

```python
"""Task 4: Histogram
Create a DataFrame with:
marks (50 random values between 0 and 100)
Plot a histogram with appropriate bins."""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create DataFrame with 50 random marks between 0 and 100
df = pd.DataFrame({
    "marks": np.random.randint(0, 101, 50)
})

# Plot histogram
plt.hist(df["marks"], bins=10)
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Histogram of Marks")
plt.show()
```

```python
"""Task 6: Value Counts Pie Chart
Create a DataFrame with:
gender (Male/Female values)
Use value_counts().
 Plot a pie chart showing gender distribution."""

import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "gender": ["Male", "Female", "Female", "Male", "Male", "Female", "Male"]
})
counts = df["gender"].value_counts()
plt.pie(counts, labels=counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
"""Task 7: Time Series Plot
Create a DataFrame with:
date (last 10 days)
temperature
Convert date to datetime."""

import pandas as pd
import matplotlib.pyplot as plt
dates = pd.date_range(end=pd.Timestamp.today(), periods=10)
temperatures = [25, 26, 24, 23, 27, 28, 26, 25, 24, 26]
df = pd.DataFrame({
    "date": dates,
    "temperature": temperatures
})
df
plt.plot(df["date"], df["temperature"])
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.title("Temperature Over Last 10 Days")
plt.xticks(rotation=45)
plt.show()
```

```python
"""Task 8: Multi-Column Plot
Create a DataFrame with:
month
revenue
expenses
Plot both revenue and expenses on the same line graph."""

import pandas as pd
import matplotlib.pyplot as plt
data = {
    "month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "revenue": [50000, 55000, 60000, 65000, 70000],
    "expenses": [30000, 32000, 35000, 36000, 40000]
}

df = pd.DataFrame(data)
plt.plot(df["month"], df["revenue"], marker='o', label="Revenue")
plt.plot(df["month"], df["expenses"], marker='o', label="Expenses")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.title("Monthly Revenue vs Expenses")
plt.legend()
plt.show()
```

```python
"""Task 9: Box Plot
Create a DataFrame with:
class
marks
Plot a box plot to compare mark distribution across classes."""

import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "class": ["A", "A", "B", "B", "C", "C"],
    "marks": [80, 85, 70, 75, 90, 95]
})
df.boxplot(column="marks", by="class")
plt.show()
```

```python
"""Task 10: Categorical Count Plot
Create a DataFrame with:
product_category
Count occurrences.
 Plot a bar chart showing product frequency.   simple code human written"""

import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
data = {
    "product_category": ["Electronics", "Clothing", "Electronics",
                         "Grocery", "Clothing", "Electronics"]
}

df = pd.DataFrame(data)
category_count = df["product_category"].value_counts()
category_count.plot(kind="bar")
plt.xlabel("Product Category")
plt.ylabel("Count")
plt.title("Product Category Frequency")
plt.show()
```

```python
"""Task 11: Cumulative Sum Plot
Create a DataFrame with:
day
daily_sales
Compute cumulative sales.
 Plot cumulative sales using a line chart."""

import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
data = {
    "day": [1, 2, 3, 4, 5],
    "daily_sales": [200, 150, 300, 250, 400]
}

df = pd.DataFrame(data)
df["cumulative_sales"] = df["daily_sales"].cumsum()
plt.plot(df["day"], df["cumulative_sales"])
plt.xlabel("Day")
plt.ylabel("Cumulative Sales")
plt.title("Cumulative Sales Over Days")
plt.show()
```

```python
"""Task 12: Data Filtering + Plot
Create a DataFrame with:
city
population
Filter cities with population greater than 1 million.
 Plot a bar chart for the filtered cities."""

import pandas as pd
import matplotlib.pyplot as plt
data = {
    "city": ["Mumbai", "Pune", "Delhi", "Kolkata", "Bangalore"],
    "population": [2000000, 900000, 1500000, 1200000, 800000]
}

df = pd.DataFrame(data)
filtered_df = df[df["population"] > 1000000]
plt.bar(filtered_df["city"], filtered_df["population"])
plt.xlabel("City")
plt.ylabel("Population")
plt.title("Cities with Population > 1 Million")
plt.show()
```
